```python
from sklearn import datasets as data
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn import datasets as data
boston = data.load_boston()
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df['value'] = boston.target
df.head()
```

```python
print(boston.DESCR)
```

```python
#Build model
x = boston.data
y = boston.target
lr = LinearRegression()
lr.fit(x,y)
```

```python
yhat = lr.predict(x)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
mean_squared_error(y,yhat)
```

```python
r2_score(y,yhat)
```

**Recursive Feature Elimination**

```python
#importing
from sklearn.feature_selection import VarianceThreshold, RFE
```

```python
columns = df.drop(['value'], axis=1).columns
```

```python
x =  boston.data
y = boston.target

lr = LinearRegression()
rfe = RFE(lr, step =1, verbose=1)
x_reduced = rfe.fit_transform(x, y)

lr.fit(x_reduced, y)
print(lr.score(x_reduced, y))
```

```python
# to check which cols are kept
columns[rfe.get_support()]
```

```python
rfe = RFE(lr, n_features_to_select =3, step=1, verbose=1)
x_red =  rfe.fit_transform(x,y)
lr.fit(x_red,y)
print(lr.score(x_red,y))
```
